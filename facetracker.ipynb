#얼굴인식 프로그래밍

```python
#모듈 설치
import cv2
import matplotlib.pyplot as plt
import cvlib as cv
```

```python
image = "city.jpg"
im = cv2.imread(image)
plt.imshow(im)
plt.show
```

```python
faces, confidences = cv.detect_face(im)
print(faces)
print(confidences)
```

```python
#얼굴인식
for face in faces:
  (startX, startY) = face[0],face[1]
  (endX, endY) = face[2], face[3]
  cv2.rectangle(im,(startX,startY),(endX,endY),(0,255,0),2)


plt.imshow(im)
plt.show()
```

```python
#결과파일저장
cv2.imwrite('result.jpg',im)
```
